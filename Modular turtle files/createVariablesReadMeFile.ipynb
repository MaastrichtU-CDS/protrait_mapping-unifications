```python
import os
import string

vrm = []

for file in os.listdir():
    if file.endswith(".ttl"):
        vrm.append(file)

#lijst van alle afzonderlijke turtle files in de folder
print(vrm)
```

```python
lines = []

for i in vrm:
    lines.append([line.rstrip('\n') for line in open("./%s" % (i))])

    #alles op 1 regel

print(lines[1])
```

```python
from more_itertools import locate
import re

for i in lines[1:]:
    #skip first element which is the prefixes file
    
    text = " ".join(i)  
    words = text.split()

    selectList = list(locate(words, lambda x: x == 'select'))
    
    for j in selectList:
        selectPart = words[j+2].translate(str.maketrans('', ''))
        print('select' + '\n' + selectPart)
        
    sparqlList = list(locate(words, lambda x: x == '?Patient'))    
    for j in sparqlList:
        sparqlPart = words[j+2].translate(str.maketrans('', ''))
        print(sparqlPart)
```
